```python
home=matches[['Home Team Name','Home Team Goals']]
home
```

```python
away=matches[['Away Team Name','Away Team Goals']]
away
```

```python
#replacing column name (Home Team Name,Home Team Goals) to (countries, Goals)&
# (Away Team Name,Away Team Goals) to (countries, Goals) so that would be able to append these columns later on 
home.columns=['countries', 'Goals']
away.columns=['countries', 'Goals']
```

```python
home
```

```python
away
```

```python
# appending these columns
goals=home.append(away,ignore_index=True)
goals
```

```python
# we can see total goal counts country wise
goals=goals.groupby('countries').sum().sort_values('Goals')
goals
```

```python
t1=matches[['Year','Home Team Name','Home Team Goals']]
t2=matches[['Year','Away Team Name','Away Team Goals']]
```

```python
# replacing the columns name
t1.columns=['Year','countries','goals']
t2.columns=['Year','countries','goals']
```

```python
records=t1.append(t2,ignore_index=True)
records
```

```python
# here performed group by so that we can understand total goals team wise
total_goals=records.groupby(['Year','countries']).sum().sort_values(['Year','goals'],ascending=False)
total_goals
```

```python
def get_label(matches):
    if matches['Home Team Goals']>matches['Away Team Goals']:
        return 'Home Team Win'
    if matches['Home Team Goals']<matches['Away Team Goals']:
        return 'Away Team Win'
    else:
        return 'Draw'
```

```python
# calling function so that we can see which team won and when there was a draw in the matches,(new columns outcomes will show this result)
matches['outcomes']=matches.apply(lambda x: get_label(x), axis=1)
```

```python
matches.head()
```

```python
# we can see total count of winners for both team type(Home team,Away team) and counts of Draw matches
out=matches['outcomes'].value_counts()
out
```

```python
# ploting winning percentage for each type team and Draw percentage 
# wecan analyse that match winning percentage for Home Team is maximum
plt.figure(figsize=(8,12))
colors=["orange","green","pink","grey"]
exp=[0,0.1,0]
out.plot.pie(autopct='%1.0f%%',colors=colors,explode=exp)
plt.title('Match Outcome')
plt.legend()
plt.show()
```
